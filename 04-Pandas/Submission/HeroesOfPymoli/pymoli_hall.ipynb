```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
purchase_csv= "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_df= pd.read_csv(purchase_csv)
```

```python
df2 = purchase_df.rename(columns = {"Purchase ID": "purchase_id", "Item ID": "item_id", "Item Name": "item_name"})
df2.head()
```

```python
#obtain player count 
player_count = purchase_df.SN.nunique()
player_count
```

```python
#create dataframe total players
total_players = pd.DataFrame({"Total Players":[player_count]})
total_players
```

```python
#calculate unique itmes
unique_items = df2.item_name.nunique()
avg_items = df2.Price.mean()
purchases = len(df2["purchase_id"].value_counts())
revenue = df2.Price.sum()
```

```python
#create dataframe
purchasing_analysis = pd.DataFrame({"unique_items" : [unique_items], "average_items": [avg_items], "total_purchases": [purchases], "total_revenue":[revenue]}) 

#add formatting
purchasing_analysis["total_revenue"] =purchasing_analysis["total_revenue"].map("${:.2f}".format)
purchasing_analysis.head()
```

```python
#get gender demographics

#males
male_people = df2.loc[df2["Gender"] == "Male", :]
males = male_people.SN.nunique()
avg_males = males / player_count* 100


#females
female_people = df2.loc[df2["Gender"] == "Female", :]
females = female_people.SN.nunique()
avg_females = females / player_count * 100


#others
other_people = df2.loc[df2["Gender"] == "Other / Non-Disclosed", :]
others = other_people.SN.nunique()
avg_others = others / player_count * 100
```

```python
#create dataframe
gender_demographics = pd.DataFrame({"total_count" : [males, females, others], "percentage_of_players": [avg_males, avg_females, avg_others]}) 

#add formatting
gender_demographics["percentage_of_players"] = gender_demographics["percentage_of_players"].map("{:.2f}%".format)
```

```python
#rename rows
gender_demographics.index = ['Males', 'Females', 'Others / Non-Disclosed']

gender_demographics
```

```python
#count purchases for males
male_purchases = len(male_people["purchase_id"].value_counts())

#count purchaes for females
female_purchases = len(female_people["purchase_id"].value_counts())

#count purcheases for "others"
others_purchases = len(other_people["purchase_id"].value_counts())

```

```python
#avg male prices
avg_male_price = male_people.Price.mean()

#avg female prices
avg_female_price = female_people.Price.mean()

#avg "others" price
avg_others_price = other_people.Price.mean()
```

```python
#total purchases for men
total_male_price = male_people.Price.sum()

#total purchases for females
total_female_price = female_people.Price.sum()

#total purchases for "others"
total_other_price = other_people.Price.sum()
```

```python
#average purchase total per male
avg_purchase_male = total_male_price / males


#average purchase total per female
avg_purchase_female = total_female_price / females

#average purcahse total per "other"
avg_purchase_other = total_other_price / others
```

```python
#create dataframe
purchasing_analysis = pd.DataFrame({"purchase_count" : [male_purchases, female_purchases, others_purchases], "average_purchase_price": [avg_male_price, avg_female_price, avg_others_price], "total_purchase_value": [total_male_price, total_female_price, total_other_price], "avg_total_purchase_per_person": [avg_purchase_male,  avg_purchase_female, avg_purchase_other]}) 
```

```python
#rename rows
purchasing_analysis.index = ['Males', 'Females', 'Others / Non-Disclosed']
purchasing_analysis

#add formatting
purchasing_analysis["average_purchase_price"] = purchasing_analysis["average_purchase_price"].map("${:.2f}".format)
purchasing_analysis["total_purchase_value"] = purchasing_analysis["total_purchase_value"].map("${:.2f}".format)
purchasing_analysis["avg_total_purchase_per_person"] = purchasing_analysis["avg_total_purchase_per_person"].map("${:.2f}".format)
purchasing_analysis
```

```python
age_bins = [0,9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9, 9999999]
age_groups = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

df2["grouped_ages"] = pd.cut(df2["Age"],age_bins, labels = age_groups, include_lowest=True)

#group by age groups
ages_df = df2.groupby("grouped_ages")
ages_df.head()
```

```python
#total players by age
total_players_age = ages_df["SN"].nunique()
#percentage of players by age
percentage_by_age = total_players_age / player_count * 100
```

```python
#age demographic table
age_data = pd.DataFrame({"percentage_by_age": percentage_by_age,"total_players_count":total_players_age})

age_data["percentage_by_age"] = age_data["percentage_by_age"].map("{:.2f}%".format)
age_data
```

```python
#conduct purchase analysis by age

#average purchase price by age
avg_age_purchase_price = ages_df.Price.mean()

#total purchase count by age
total_purchase_count_byage= ages_df["item_name"].count()

#total purchase amount by age
age_purchase_total = ages_df.Price.sum()

#average purchase per age
avg_purchase_per_age = age_purchase_total/ total_players_age
```

```python
#age analysis dataframe
age_purchase_analysis = pd.DataFrame({"total_purchase_count": total_purchase_count_byage,"average_purchase_price": avg_age_purchase_price, "total_purchase_value": age_purchase_total,"avg_purchase_total_per_person": avg_purchase_per_age})
                    
#add formatting
age_purchase_analysis["average_purchase_price"] = age_purchase_analysis["average_purchase_price"].map("${:.2f}".format)
age_purchase_analysis["total_purchase_value"] = age_purchase_analysis["total_purchase_value"].map("${:.2f}".format)
age_purchase_analysis["avg_purchase_total_per_person"] = age_purchase_analysis["avg_purchase_total_per_person"].map("${:.2f}".format)
age_purchase_analysis
```

```python
#group by Screen Name
df4 = df2.groupby("SN")
```

```python
#calculate purcahse countby screen name
purchase_count = df4.item_id.count()

#calculate average purchase price by screen name
average_purchase_price = df4.Price.mean()

#calculate total purchase value
total_purchase_value = df4.Price.sum().sort_values(ascending=False)
```

```python
#create purchase analysis dataframe
spenders = pd.DataFrame({"purchase_count":purchase_count, "avg_purchase_price" : average_purchase_price, "total_purchase_value": total_purchase_value})

#sort by total purchase value
top_5 = spenders.sort_values(["total_purchase_value"], ascending=False).head()

#add formatting
top_5["avg_purchase_price"] = top_5["avg_purchase_price"].map("${:.2f}".format)
top_5["total_purchase_value"] = top_5["total_purchase_value"].map("${:.2f}".format)
top_5
```

```python
#group by item id and item name
df5 = df2.groupby(["item_id","item_name"])
```

```python
#calculate purchase count
purchase_count2 = df5.item_id.count().sort_values(ascending=False)
```

```python
#calculate purchase value
purchase_value = df5.Price.sum()

#calculate price of individual item
price_of_item = purchase_value / purchase_count2
```

```python
#create dataframe on popular items
popular_items = pd.DataFrame({"purchase_count": purchase_count2, "item_price": price_of_item, "total_purchase_value": purchase_value})

#display top 5
top_5_items = popular_items.sort_values(["purchase_count"], ascending=False).head()

#add formatting
top_5_items["item_price"] = top_5_items["item_price"].map("${:.2f}".format)
top_5_items["total_purchase_value"] = top_5_items["total_purchase_value"].map("${:.2f}".format)

top_5_items
```

```python
#sort top 5 items by total purchase value to get the information
top_5_most_profitable = popular_items.sort_values(["total_purchase_value"],ascending=False).head()

#add formatting
top_5_most_profitable["item_price"] = top_5_most_profitable["item_price"].map("${:.2f}".format)
top_5_most_profitable["total_purchase_value"] = top_5_most_profitable["total_purchase_value"].map("${:.2f}".format)

top_5_most_profitable
```
