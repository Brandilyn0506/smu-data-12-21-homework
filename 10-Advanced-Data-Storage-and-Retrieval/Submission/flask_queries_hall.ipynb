```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine, inspect
```

```python
 path = "sqlite:///Resources/hawaii.sqlite"
engine = create_engine(path)
conn = engine.connect()
```

```python
 inspector = inspect(engine)

tables = inspector.get_table_names()
tables
```

```python
query = """
            SELECT date, station, prcp
            FROM measurement
            ORDER BY date asc, station asc
            """

df = pd.read_sql(query, conn)
df
```

```python
query = """
            SELECT *
            FROM station
            """

df2 = pd.read_sql(query, conn)
```

```python
df2['station'].values.tolist()
```

```python
query = """
        SELECT date, tobs
        FROM measurement
        WHERE station = 'USC00516128' AND date >= '2016-08-23'
        """
df3 = pd.read_sql(query, conn)
df3
```

```python
df3['tobs'].values.tolist()
```

```python
query= """
        SELECT AVG(tobs) AS "Average temp"
        FROM measurement
        WHERE date = '2016-06-01';
        """
df4 = pd.read_sql(query, conn)
df4
```

```python
query= """
        SELECT AVG(tobs) AS "Average temp",MAX(tobs) AS "Max temp", MIN(tobs) AS "Min temp"
        FROM measurement
        WHERE date = '2016-06-01';
        """
df4 = pd.read_sql(query, conn)
```
